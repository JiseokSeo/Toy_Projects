```python
def rle_decode(mask_rle, shape):
    '''
    mask_rle: run-length as string formated (start length)
    shape: (height,width) of array to return 
    Returns numpy array, 1 - mask, 0 - background

    '''
    s = mask_rle.split()
    starts, lengths = [np.asarray(x, dtype=int) for x in (s[0:][::2], s[1:][::2])]
    starts -= 1
    ends = starts + lengths
    img = np.zeros(shape[0]*shape[1], dtype=np.uint8)
    for lo, hi in zip(starts, ends):
        img[lo:hi] = 1
    return img.reshape(shape)  # Needed to align to RLE direction
```

```python
import os
import SimpleITK as sitk
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display
```

```python
masks = []
for k in range(len(case123_day20)):
  mask = rle_decode(case123_day20['segmentation'][k], (266, 266))
  masks.append(mask)

mask_3d = np.stack(masks, axis=0)
sitk.WriteImage(sitk.GetImageFromArray(mask_3d), '/content/case123_day20_stomach.nii.gz')
```

```python
# 이미지 읽기
segmented_3d_image = sitk.ReadImage('/content/case123_day20_stomach.nii.gz')

# 이미지를 NumPy 배열로 변환
segmented_3d_image_arr = sitk.GetArrayFromImage(segmented_3d_image)

# 슬라이더를 사용하여 다양한 축면의 이미지를 시각화
def plot_slice(slice_idx):
    plt.figure(figsize=(6, 6))
    plt.imshow(segmented_3d_image_arr[slice_idx, :, :], cmap="gray")
    plt.axis("off")
    plt.show()

slice_slider = widgets.IntSlider(min=0, max=segmented_3d_image_arr.shape[0]-1, step=1, value=0)
widgets.interact(plot_slice, slice_idx=slice_slider)
```

```python
import os
import SimpleITK as sitk
import numpy as np
import plotly.graph_objs as go

# 이미지 읽기
segmented_3d_image = sitk.ReadImage('/content/case123_day20_stomach.nii.gz')

# 이미지를 NumPy 배열로 변환
segmented_3d_image_arr = sitk.GetArrayFromImage(segmented_3d_image)

# 이미지를 이진 마스크로 변환 (예: 1의 값을 가진 영역만 시각화)
binary_mask = (segmented_3d_image_arr == 1)

# 입체 구조 시각화를 위한 x, y, z 좌표 추출
z, y, x = np.where(binary_mask)

# 간격 설정
spacing = (3, 1, 1)
z = z * spacing[0]
y = y * spacing[1]
x = x * spacing[2]

# Plotly 3D Scatter Plot
scatter_plot = go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(
        size=3,
        opacity=0.5
    )
)

layout = go.Layout(scene=dict(aspectmode='data'))
fig = go.Figure(data=[scatter_plot], layout=layout)
fig.show()
```
